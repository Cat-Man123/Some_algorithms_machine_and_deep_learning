```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
iris = load_iris()
print(iris.keys())
```

```python
print(iris["target"])
print(iris["target_names"])
print(iris["feature_names"])
```

```python
log_reg1 = LogisticRegression(C=1, max_iter=1000)
log_reg10 = LogisticRegression(C=10, max_iter=1000)
log_reg100 = LogisticRegression(C=100, max_iter=1000)
log_reg1000 = LogisticRegression(C=1000, max_iter=1000)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris["data"], iris["target"], random_state=0)
```

```python
log_reg1.fit(X_train, y_train)
log_reg10.fit(X_train, y_train)
log_reg100.fit(X_train, y_train)
log_reg1000.fit(X_train, y_train)
```

```python
print(f"Logistic regression with C=1 training score: {log_reg1.score(X_train, y_train)}")
print(f"Logistic regression with C=1 test score: {log_reg1.score(X_test, y_test)}")

print(f"Logistic regression with C=10 training score: {log_reg10.score(X_train, y_train)}")
print(f"Logistic regression with C=10 test score: {log_reg10.score(X_test, y_test)}")

print(f"Logistic regression with C=100 training score: {log_reg100.score(X_train, y_train)}")
print(f"Logistic regression with C=100 test score: {log_reg100.score(X_test, y_test)}")

print(f"Logistic regression with C=1000 training score: {log_reg1000.score(X_train, y_train)}")
print(f"Logistic regression with C=1000 test score: {log_reg1000.score(X_test, y_test)}")
```

```python
pred_new = np.array([[5.6, 2.2, 1.0, 1.2]])
prediction_log1 = log_reg1.predict(pred_new)
prediction_log10 = log_reg10.predict(pred_new)
prediction_log100 = log_reg100.predict(pred_new)
prediction_log1000 = log_reg1000.predict(pred_new)
```

```python
print(f"Prediction log 1: {prediction_log1}")
print(f"Prediction log 10: {prediction_log10}")
print(f"Prediction log 100: {prediction_log100}")
print(f"Prediction log 1000: {prediction_log1000}")

print(iris["target_names"][prediction_log1])
print(iris["target_names"][prediction_log10])
print(iris["target_names"][prediction_log100])
print(iris["target_names"][prediction_log1000])
```
