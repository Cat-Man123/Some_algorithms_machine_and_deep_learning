```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
diabetes = load_diabetes()
print(diabetes.keys())
```

```python
print(diabetes["data"])
print(diabetes["target"])
print(diabetes["feature_names"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(diabetes["data"], diabetes["target"], random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
li_reg = LinearRegression()
li_reg.fit(X_train, y_train)
```

```python
print(f"Training score: {li_reg.score(X_train, y_train)}")
print(f"Test score: {li_reg.score(X_test, y_test)}")
print(f"Coef: {li_reg.coef_}")
print(f"Intercept: {li_reg.intercept_}")
```

```python
X_pred = np.array([[0.2, 2.0, 1.0, 0.5, 0.2, 0.001, 0.6, 2.0, 1.0, 0.4]])
prediction = li_reg.predict(X_pred)
print(f"Prediction: {prediction}")
```
